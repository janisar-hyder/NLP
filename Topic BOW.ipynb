```python
import os
import spacy
nlp=spacy.load("en_core_web_sm")
from autocorrect import Speller
spell = Speller()
import nltk
from nltk import sent_tokenize
path='D:\\NLP' 
filepath=os.listdir(path)

D_F=[]
def filetext(filename):
    for file in filename:
        with open(os.path.join(path, file), "r", encoding="utf-8", errors="ignore") as f:
            text = f.read() 
            to_store_sentences = []
            sent_ = sent_tokenize(text)
            for sentence in sent_:
                to_store_sentences.append(sentence)
            index = [text, type(text), to_store_sentences] 
            D_F.append(index)

filetext(filepath)



D_S_P=[]
for sentence in D_F:
    D_S_P.append(sentence[2])

def PreProcessingspacy(text):
    text=spell(text)
    token=[]
    tokens=nlp(text)
    token=[tokens]
    pos=[]
    for w in tokens:
        pos.append(w.text)
        pos.append(w.pos_)
    token_without_stopwords = [w.text for w in tokens if not  w.is_stop]    
    token_in_lowercase = [w.lower() for w in token_without_stopwords]
    lemmas = [w.lemma_ for w in tokens if not (w.is_stop or not w.is_alpha)]
    lemmas=[w.lower() for w in lemmas]
    
    return [token,pos,token_without_stopwords,token_in_lowercase,lemmas]

i=0
j=0
for file in D_S_P:
    for sentence in file:
        data=PreProcessingspacy(sentence)
        sentence=[sentence,data]   
        D_S_P[i][j]=sentence
        j=j+1
    j=0
    i=i+1


import pickle
save1=open("PreProcessed_Data.pickle","wb")
pickle.dump(D_S_P,save1)
save1.close()


import pickle
file3=open("PreProcessed_Data.pickle","rb")
D_S_P=pickle.load(file3)

BOW=[]
for index in range(0,len(D_S_P)):
    for index1 in range(0,len(D_S_P[index])):
        lemmas=D_S_P[index][index1][-1][-1]
        for l in lemmas:
            BOW.append(l)

print(BOW)

Freuency={}
for word in BOW:
    if word in Freuency:
        Freuency[word]=Freuency[word]+1
    else:
        Freuency[word]=1


listed=list(Freuency.items())

def insertionsort(Freuency):
    for i in range(1, len(Freuency)):
        valuetochange=Freuency[i]
        print(valuetochange,end="|")
        a=i-1
        while a >= 0 and Freuency[a][1] < valuetochange[1]:
            Freuency[a+1]=Freuency[a]
            a=a-1
        Freuency[a+1]=valuetochange

insertionsort(listed)
sortedd = dict(listed)

print(sortedd)
top50=[listed[i] for i in range(0,50)]
top50sortedd=dict(top50)
print(len(top50sortedd))
print(top50sortedd)
```

## `BASIC IMPORTS and File Reading`

```python
import os
import spacy
import pickle
from autocorrect import Speller
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
nlp = spacy.load("en_core_web_sm")
spell_Check = Speller()
import random

path = 'D:\\NLP\\' 
filepath = os.listdir(path)

short_pos = open("D:\\NLP\\pos.txt", "r").read()
short_neg = open("D:\\NLP\\neg.txt", "r").read()

documents = []

for r in short_pos.split('\n'):
    documents.append((r, "pos"))

for r in short_neg.split('\n'):
    documents.append((r, "neg"))

print(documents[:5])
print(len(documents))
```

## `Preprocessing and Frequency Distribution`

```python

all_words = []

def preprocess_text(text):
    text = spell_Check(text)
    tokens = nlp(text)
    tokens = [t.text for t in tokens if t.is_alpha and not t.is_stop]
    tl = [t.lower() for t in tokens]
    lemmas = [tl.lemma_ for tl in nlp(" ".join(tl))]
    return lemmas

processed_pos = preprocess_text(short_pos)
processed_neg = preprocess_text(short_neg)
print("Positive: ",len(processed_pos))
print("Negative: ",len(processed_neg ))

for w in processed_pos:
    all_words.append(w)
for w in processed_neg :
    all_words.append(w)
print("Total before FreqDisttribution:",len(all_words))
all_words = nltk.FreqDist(all_words)
print("Total After FreqDisttribution:",len(all_words))
```

## `Pickling`

```python
newfile=open("Featuressss.pickle","wb")
pickle.dump(all_words,newfile)
newfile.close()

myfile=open("Featuressss.pickle","rb")
all_words=pickle.load(myfile)
```

## `Printing Keys After Preprocessing`

```python
word_features = list(all_words.keys())
print(len(word_features) , word_features)
```

## `Finding Features (True/False)`

```python
def find_features(document):
    words = word_tokenize(document)
    features = {}
    for w in word_features:
        features[w] = (w in words)

    return features


featuresets = [(find_features(rev), category) for (rev, category) in documents]

random.shuffle(featuresets)
print("Number of datasets:", len(featuresets))
featuresets[0]
```
